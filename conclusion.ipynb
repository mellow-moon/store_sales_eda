# Вывод

### 1. Какая доля продаж приходится на каждый тип продукта? У каких продуктов продажи больше и у каких меньше?

Доли продаж распределены не равномерно. Наибольшую долю занимают Товары первой необходимости (32%), Напитки (20%) и Сельскохозяйственная продукция (11%). Наименьшую - Книги (<1%), Уход за детьми (<1%), Бытовая техника (<1%).

### 2. Какая доля продаж приходится на каждый магазин? В каких магазинах наибольший объем продаж? Что это за магазины?

Наибольшая доля продаж приходится на магазины 44 (5.78%), 45 (5.08%) и 47 (4.75%).

Большинство магазинов с лучшими продажами находятся в городе Quito, при этом среди лучших магазинов наиболее часто встречаются магазины типа A.

Из 10 магазинов с самыми низкими продажами 4 находятся в городе Guayaquil. Половина магазинов с низкими продажами имеют тип C.

### 4. Какая доля продаж приходится на каждый штат?

Доли продаж по каждому штату распределены не равномерно. Наибольшую долю занимают штаты Pichincha (54%), Guayas (15%) и Azuay (5%).

### 5. Оказывают ли цены на нефть влияние на объем продаж?

В данных присутствует отрицательная корреляция между ценами на нефть и суммарным объемом продаж.

### 7. В каком штате наибольшее количество клиентов?

Наибольший объем продаж среди всех штатов показал штат Pichincha, это позволяет сделать вывод, что в этом штате проживает наибольшее количество клиентов.

### 8. В каких месяцах продажи были наибольшими и наименьшими?

Рекорд по продажам был установлен в декабре 2016 года. Декабрь - это самый доходный месяц в году (кроме 2017 года, так как данные о продажах в этом году были собраны только до августа).

С каждым годом объем продаж увеличивается.

В 2014 году есть два месяца с аномально высокими продажами.

В июне 2015 года произошел сильный рост продаж, затем рост продолжился, но уже более медленными темпами.
